```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img_path="/content/download.jpg"
img=cv2.imread(img_path)
plt.imshow(img[:, :, ::-1])
```

```python
face=DeepFace.detectFace(img_path,
                         target_size=(224,224),
                         detector_backend='opencv'
)
```

```python
plt.imshow(face[0])
```

```python
obj=DeepFace.analyze(img_path)
```

```python
emotion=obj['dominant_emotion']
age=obj['age']
gender=obj['dominant_gender']


print(f'emotion: {emotion}')
print(f'age: {age}')
print(f'gender: {gender}')
```
